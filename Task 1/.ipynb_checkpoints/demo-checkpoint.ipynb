<h3>Import Libraries and Load Dataset</h3>

```python
import numpy as np
import matplotlib.pyplot as plt
from task1 import MnistClassifier
from PIL import Image
import tensorflow as tf
```

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
x_test_sample = x_test[1]
plt.imshow(x_test_sample, cmap='gray')
plt.show()

image_pil = Image.fromarray((x_test_sample * 255).astype(np.uint8))
```

<h3>Random Forest</h3>

```python
image_pil = Image.fromarray((x_test_sample * 255).astype(np.uint8))

model_kwargs = {"n_estimators" : 100}
clf = MnistClassifier("rf", model_kwargs)
clf.train(x_train, y_train)
prediction = clf.predict(image_pil)
print("Predicted class:", prediction)
```

```python
x_test_samples = x_test[:9]

clf = MnistClassifier("rf")

fig, axes = plt.subplots(3, 3, figsize=(8, 8))

for i, ax in enumerate(axes.flat):
    image_pil = Image.fromarray((x_test_samples[i] * 255).astype(np.uint8))
    
    prediction = clf.predict(image_pil)
    
    ax.imshow(x_test_samples[i], cmap='gray')
    ax.set_title(f"Pred: {prediction}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```

<h3>FeedForwardNeuralNetwork</h3>

```python
x_test_sample = x_test[4]
plt.imshow(x_test_sample, cmap='gray')
plt.show()
```

```python
image_pil = Image.fromarray((x_test_sample * 255).astype(np.uint8))

clf = MnistClassifier("nn")
clf.train(x_train, y_train)
prediction = clf.predict(image_pil)
print("Predicted class:", prediction)
```

```python
x_test_samples = x_test[:9]

clf = MnistClassifier("nn")

fig, axes = plt.subplots(3, 3, figsize=(8, 8))

for i, ax in enumerate(axes.flat):
    image_pil = Image.fromarray((x_test_samples[i] * 255).astype(np.uint8))
    
    prediction = clf.predict(image_pil)
    
    ax.imshow(x_test_samples[i], cmap='gray')
    ax.set_title(f"Pred: {prediction}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
